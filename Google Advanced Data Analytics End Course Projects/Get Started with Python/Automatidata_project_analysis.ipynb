## Loading Data of Taxi and Limousine Commission For Analysis

```python
import pandas as pd

data_df: pd.DataFrame = pd.read_csv("2017_Yellow_Taxi_Trip_Data (1).csv")
```

## Data Cleaning and Preparation

```python
data_df.info()
```

### The Two Columns Which is
### 1. tpep_pickup_datetime
### 2. tpep_dropoff_datetime
### These two column data type is in object but they both needs to convert with acutall date object type
### So I Convert to Date time Data Type

```python
data_df['tpep_pickup_datetime'] = pd.to_datetime(data_df['tpep_pickup_datetime'])
data_df['tpep_dropoff_datetime'] = pd.to_datetime(data_df['tpep_dropoff_datetime'])
```

## Understand the data - Investigate the variables

```python
data_df_sort = data_df.sort_values(by="trip_distance",ascending=False)
data_df_sort.head(10)
```

```python
total_amount_sorted = data_df_sort.sort_values(by="total_amount",ascending=False)[['total_amount']]
total_amount_sorted.head(20)
```

```python
total_amount_sorted.tail(20)
#The values align with our earlier data discovery, where we noticed that the longest rides are approximately 33 miles.
# Yes, the first two values are significantly higher than the others.
#The most expensive rides are not necessarily the longest ones.
```

### How many of each payment type are represented in the data?


```python
data_df['payment_type'].value_counts()
```

```python
payment_type_percentage = data_df['payment_type'].value_counts(normalize=True)*100
payment_type_percentage
```

### Creating a Graph for better understanding

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.barplot(x=payment_type_percentage.index, y=payment_type_percentage.values)
plt.xlabel("Payment Types")
plt.ylabel("Percentage")
plt.show()
```

### Calculating Average tips of Credit Card = 1 and Cash = 2

```python
# What is the average tip for trips paid for with credit card?
avg_tip_cc = data_df[data_df['payment_type']==1]['tip_amount'].mean()
print(avg_tip_cc)

# What is the average tip for trips paid for with cash?
avg_tip_cash = data_df[data_df['payment_type'] == 2]['tip_amount'].mean()
print(avg_tip_cash)
```

### # How many times is each vendor ID represented in the data?

```python
data_df['VendorID'].value_counts()
```

```python
percentage_of_vendorId = data_df['VendorID'].value_counts(normalize=True)*100
percentage_of_vendorId
```

### Creating Bar Plot For Better understanding

```python
sns.barplot(x=percentage_of_vendorId.index, y=percentage_of_vendorId.values,color='green')
plt.xlabel("VendorID")
plt.ylabel("Percentage")
plt.show()
```

### What is the mean total amount for each vendor?


```python
data_df.groupby(['VendorID']).mean(numeric_only=True)[['total_amount']]
```

### Filter the data for credit card payments only


```python
credit_card = data_df[data_df['payment_type'] == 1]
credit_card['passenger_count'].value_counts()
```

### # Calculate the average tip amount for each passenger count (credit card payments only)


```python
credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```
