```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
user_names = ["user_id", "gender", "age", "occupation", "zip"]
users = pd.read_table("users.dat", sep="::",
                      header=None, names=user_names, engine="python")

ratings_names = ["user_id", "movie_id", "rating", "timestamp"]
ratings = pd.read_table("ratings.dat", sep="::",
                        header=None, names=ratings_names, engine="python")

movie_names = ["movie_id", "title", "genres"]
movies = pd.read_table("movies.dat", sep="::",
                       header=None, names=movie_names, engine="python")

```

### Verify The Data Frame Of Users, ratings, movies

```python
users.head()
```

```python
ratings.head()
```

```python
movies.head()
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
data.head()
```

```python
data.info()
```

```python
data.iloc[1]
```

```python
mean_ratings= data.pivot_table("rating", index="title",
                      columns="gender",aggfunc="mean")

mean_ratings
```

```python
with_group_by = data.groupby(['title','gender'])[['rating']].mean()
with_group_by
```

```python
ratings_by_title = data.groupby("title").size()
ratings_by_title.head()
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
active_titles
```

```python
mean_ratings = mean_ratings.loc[active_titles]
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by="F", ascending=False)
top_female_ratings
```

## Measuring Rating Disagreement


```python
mean_ratings["diff"] = mean_ratings["M"] - mean_ratings["F"]
sorted_by_diff = mean_ratings.sort_values("diff")
sorted_by_diff.head()
```

```python
sorted_by_diff[::-1]
```

## Now We check the  most disagreement among viewers, independent of gender identification

```python
rating_std_by_title = data.groupby(["title"])[['rating']].std()
rating_std_by_title = rating_std_by_title.loc[active_titles]
rating_std_by_title.head()
```

```python
rating_std_by_title.sort_values(by="rating",ascending=False)[:10]
```

```python
movies["genre"] = movies.pop("genres").str.split("|")
movies.head()
```

### Now Merge all three tables again with movies_exploded

```python
ratings_with_genre = pd.merge(pd.merge(movies_exploded, ratings), users)
ratings_with_genre.head()
```
